Generally, we solve the Reaction diffusion eq $D\nabla^2 \phi_{out} + s(\phi_{out}) \approx 0$ for $\phi_{out}$ in a single annular shell sector.

$s(\phi_{out})$ can be approximated as a linear function in $\phi_{out}$ as:

$s(\phi_{out}) = A - B \phi_{out}$ Note that $B > 0$ has to be positive, else the solution will diverge. 

$s[\phi_{out} \vert_{r=R} ] = s(\phi^{eq}_{out}) \mathrm{~~~and~~~} s[\phi_{out} \vert_{r=R+L}] = s(\phi_{shell})$

\\\\

Solving for $A, B$ gives $B = \left ( \frac{s(\phi_{shell}) - s(\phi^{eq}_{out})}{\phi^{eq}_{out} - \phi_{shell}} \right )$ and $A = s(\phi_{shell}) + B \phi_{shell}$

Hence, we solve the equation $D\nabla^2 \phi_{out} +  A - B \phi_{out} = 0 \cdot \cdot \cdot \cdot \cdot (1)$

The boundary conditions for Eq. (1) are $\phi_{out} \vert_{r=R} = \phi^{eq}_{out},~ \phi_{out} \vert_{r=R+L} = \phi_{shell}$

### Analytical solution

```python
Image("solution3D.png", width = 600)
```

Considering a spherical co-ord system centered at the centre of each droplet, since $\Delta s$ is small enough, we essentially solve Eq. (1) has only a radial dependance.

\\\\

Then Eq. (1) for 3-D case is $D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{2}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$, for which the analytical solution is $\phi_{out} = \frac{A}{B} + \frac{C_1~e^{-\frac{\sqrt{B}~r}{\sqrt{D}}}}{r} + \frac{C_2 \sqrt{D}~e^{\frac{\sqrt{B}~r}{\sqrt{D}}}}{\sqrt{B}~r}$

#### Initialize the variables required

```python
r, theta, phi = sym.symbols('r, theta, phi', positive = True)

B, R, L, D, phiEqout = sym.symbols('B, R, L, D, \phi^{eq}_{out}', positive = True)

A, phishell, SphiEqout, Sphishell = sym.symbols('A, \phi_{shell}, s(\phi^{eq}_{out}), s(\phi_{shell})', real = True)
```

######################################################################################################################

```python
C_1, C_2 = sym.symbols('C_1, C_2', real = True)
```

```python
phi_out = A/B + (C_1/r)*(exp(-sqrt(B)*r/sqrt(D))) + (C_2/r)*(sqrt(D)/sqrt(B))*(exp(sqrt(B)*r/sqrt(D)))

simplify(phi_out)
```

### Verify the solution by plugging it in the equation:

$D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{2}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$

```python
Image("laplacian3D.jpeg", width = 500)
```

```python
del_phi_out_by_del_r = diff(phi_out, r)
```

```python
del_phi_out_by_del_phi = diff(phi_out, phi)
```

```python
del_phi_out_by_del_theta = diff(phi_out, theta)
```

```python
spherical_laplacian = (1/(r*r))*diff((del_phi_out_by_del_r*(r*r)), r) + \
                        (1/(r*r*sin(phi)*sin(phi)))*diff(del_phi_out_by_del_theta, theta) + \
                        (1/(r*r*sin(phi)))*diff((del_phi_out_by_del_phi*sin(phi)), phi)
```

```python
if (simplify(D * spherical_laplacian + A - B*phi_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

### Solve for $C_1$ and $C_2$

```python
phi_out
```

Use BC $\#1 \longrightarrow \phi_{out} \vert_{r=R} = \phi^{eq}_{out}$

```python
eq1 = Eq( phi_out.subs({r:(R)}) - phiEqout, 0)

eq1
```

######################################################################################################################

Use BC $\#2 \longrightarrow \phi_{out} \vert_{r=R+L} = \phi_{shell}$

```python
eq2 = Eq( phi_out.subs({r:(R + L)}) - phishell, 0)

eq2
```

```python
solution = list(solve((eq1, eq2), (C_1, C_2)).values())

solved_C_1, solved_C_2 = solution[0], solution[1]
```

######################################################################################################################

```python
solved_phi_out = phi_out.subs({C_1: solved_C_1, C_2: solved_C_2})
```

### Verify the solution by plugging it in the equation (again):

$D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{2}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$

```python
Image("laplacian3D.jpeg", width = 500)
```

```python
del_solved_phi_out_by_del_r = diff(solved_phi_out, r)
```

```python
del_solved_phi_out_by_del_phi = diff(solved_phi_out, phi)
```

```python
del_solved_phi_out_by_del_theta = diff(solved_phi_out, theta)
```

```python
spherical_laplacian = (1/(r*r))*diff((del_solved_phi_out_by_del_r*(r*r)), r) + \
                        (1/(r*r*sin(phi)*sin(phi)))*diff(del_solved_phi_out_by_del_theta, theta) + \
                        (1/(r*r*sin(phi)))*diff((del_solved_phi_out_by_del_phi*sin(phi)), phi)
```

```python
if (simplify(D * spherical_laplacian + A - B*solved_phi_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

### Fluxes evaluated outside the droplet $=$ in the sector

```python
Image("gradient3D.jpeg", width = 300)
```

```python
J_OUT = -D*(diff(solved_phi_out, r) + (1/r)*diff(solved_phi_out, phi) + (1/(r*sin(phi)))*diff(solved_phi_out, theta))

integrated_flux = integrate(J_OUT * r*r * sin(phi), (phi, 0, pi), (theta, 0, 2*pi))

integrated_flux_at_the_surface = integrated_flux.subs({r:R})
```

```python
simplify(integrated_flux_at_the_surface)
```

######################################################################################################################

\\\\

In the above expression, $B = \left ( \frac{s(\phi_{shell}) - s(\phi^{eq}_{out})}{\phi^{eq}_{out} - \phi_{shell}} \right )$ and $A = s(\phi_{shell}) + B \phi_{shell}$

```python
# B = ((Sphishell - SphiEqout)/(phiEqout - phishell))

# A = Sphishell + B*phishell
```

## Passive Droplet check $\longrightarrow A, B = 0$

```python
integrated_flux_at_the_surface_without_reactions = limit(limit(integrated_flux_at_the_surface, A, 0), B, 0)

integrated_flux_at_the_surface_without_reactions
```

```python
expression_in_code_without_reactions = 4 * pi * D * R * (1 + R / L) * (phiEqout - phishell)

expression_in_code_without_reactions
```

```python
if (simplify(integrated_flux_at_the_surface_without_reactions - expression_in_code_without_reactions) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```
