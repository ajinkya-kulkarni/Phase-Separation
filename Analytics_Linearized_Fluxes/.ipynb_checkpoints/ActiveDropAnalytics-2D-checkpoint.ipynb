Generally, we solve the Reaction diffusion eq $D\nabla^2 \phi_{out} + s(\phi_{out}) \approx 0$ for $\phi_{out}$ in a single annular shell sector.

$s(\phi_{out})$ can be approximated as a linear function in $\phi_{out}$ as:

$s(\phi_{out}) = A - B \phi_{out}$ Note that $B > 0$ has to be positive, else the solution will diverge. 

$s[\phi_{out} \vert_{r=R} ] = s(\phi^{eq}_{out}) \mathrm{~~~and~~~} s[\phi_{out} \vert_{r=R+L}] = s(\phi_{shell})$

\\\\

Solving for $A, B$ gives $B = \left ( \frac{s(\phi_{shell}) - s(\phi^{eq}_{out})}{\phi^{eq}_{out} - \phi_{shell}} \right )$ and $A = s(\phi_{shell}) + B \phi_{shell}$

Hence, we solve the equation $D\nabla^2 \phi_{out} +  A - B \phi_{out} = 0 \cdot \cdot \cdot \cdot \cdot (1)$

The boundary conditions for Eq. (1) are $\phi_{out} \vert_{r=R} = \phi^{eq}_{out},~ \phi_{out} \vert_{r=R+L} = \phi_{shell}$

### Analytical solution

```python
Image("solution2D_active.png", width = 600)
```

Considering a spherical co-ord system centered at the centre of each droplet, since $\Delta s$ is small enough, we essentially solve Eq. (1) has only a radial dependance.

\\\\

Then Eq. (1) for 2-D case is $D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{1}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$, for which the analytical solution is $\phi_{out} = \frac{A}{B} + C_1 \mathrm{~BesselJ}_0 ~ (i\sqrt{\frac{B}{D}} r) + C_2 \mathrm{~BesselY}_0 ~ (-i\sqrt{\frac{B}{D}} r)$

#### Initialize the variables required

```python
r, theta = sym.symbols('r, theta', positive = True)

B, R, L, D, phiEqout = sym.symbols('B, R, L, D, \phi^{eq}_{out}', positive = True)

A, phishell, SphiEqout, Sphishell = sym.symbols('A, \phi_{shell}, s(\phi^{eq}_{out}), s(\phi_{shell})', real = True)
```

######################################################################################################################

```python
C_1, C_2 = sym.symbols('C_1, C_2', real = True)
```

```python
phi_out = A/B + C_1 * besselj(0, r * sqrt(-B/D)) + C_2 * bessely(0, -r * sqrt(-B/D))

simplify(phi_out)
```

### Verify the solution by plugging it in the equation:

$D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{1}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_phi_out_by_del_r = diff(phi_out, r)
```

```python
del_phi_out_by_del_theta = diff(phi_out, theta)
```

```python
del_2_phi_out_by_del_theta_2 = diff(del_phi_out_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_phi_out_by_del_r), r) + (1/(r*r))*del_2_phi_out_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + A - B*phi_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

### Solve for $C_1$ and $C_2$

```python
phi_out
```

Use BC $\#1 \longrightarrow \phi_{out} \vert_{r=R} = \phi^{eq}_{out}$

```python
eq1 = Eq( phi_out.subs({r:(R)}) - phiEqout, 0)

eq1
```

######################################################################################################################

Use BC $\#2 \longrightarrow \phi_{out} \vert_{r=R+L} = \phi_{shell}$

```python
eq2 = Eq( phi_out.subs({r:(R + L)}) - phishell, 0)

eq2
```

```python
solution = list(solve((eq1, eq2), (C_1, C_2)).values())

solved_C_1, solved_C_2 = solution[0], solution[1]
```

```python
solved_phi_out = phi_out.subs({C_1: solved_C_1, C_2: solved_C_2})
```

```python
simplify(solved_phi_out)
```

######################################################################################################################

### Verify the solution by plugging it in the equation (again):

$D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{1}{r}\frac{\partial \phi_{out}}{\partial r} \right ] + A - B \phi_{out} = 0$

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_solved_phi_out_by_del_r = diff(solved_phi_out, r)
```

```python
del_solved_phi_out_by_del_theta = diff(solved_phi_out, theta)
```

```python
del_2_solved_phi_out_by_del_theta_2 = diff(del_solved_phi_out_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_solved_phi_out_by_del_r), r) + \
                    (1/(r*r))*del_2_solved_phi_out_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + A - B*solved_phi_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

######################################################################################################################

### Fluxes evaluated outside the droplet $=$ in the sector

```python
Image("gradient2D.jpeg", width = 300)
```

```python
J_OUT = -D*( diff(solved_phi_out, r) + (1/r)*diff(solved_phi_out, theta))

integrated_flux = integrate(J_OUT * r, (theta, 0, 2*pi))

integrated_flux_at_the_surface = integrated_flux.subs({r:R})
```

```python
simplify(integrated_flux_at_the_surface)
```

## Passive Droplet check $\longrightarrow A, B = 0$

Generally, we solve the Diffusion eq $D\nabla^2 \phi_{out}) \approx 0$ for $\phi_{out}$ in a single annular shell sector.

Hence, we solve the equation $D\nabla^2 \phi_{out} = 0 \cdot \cdot \cdot \cdot \cdot (1)$

The boundary conditions for Eq. (1) are $\phi_{out} \vert_{r=R} = \phi^{eq}_{out},~ \phi_{out} \vert_{r=R+L} = \phi_{shell}$

### Analytical solution

```python
Image("solution2D_passive.png", width = 600)
```

Considering a spherical co-ord system centered at the centre of each droplet, since $\Delta s$ is small enough, we essentially solve Eq. (1) has only a radial dependance.

\\\\

Then Eq. (1) for 2-D case is $D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{1}{r}\frac{\partial \phi_{out}}{\partial r} \right ] = 0$, for which the analytical solution is $\phi_{out} = C_3 \log(r) + C_4$

```python
C_3, C_4 = sym.symbols('C_3, C_4', real = True)
```

```python
phi_out_passive = C_3 * log(r) + C_4

simplify(phi_out_passive)
```

### Verify the solution by plugging it in the equation:

$D\left [ \frac{\partial^2 \phi_{out}}{\partial r^2} + \frac{1}{r}\frac{\partial \phi_{out}}{\partial r} \right ] = 0$

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_phi_out_passive_by_del_r = diff(phi_out_passive, r)
```

```python
del_phi_out_passive_by_del_theta = diff(phi_out_passive, theta)
```

```python
del_2_phi_out_passive_by_del_theta_2 = diff(del_phi_out_passive_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_phi_out_passive_by_del_r), r) + (1/(r*r))*del_2_phi_out_passive_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```
