```python
N_DROPLETS = int(1e2)
```

```python
single_droplet_timestamps_1D = np.load('time_1D.npy')

single_droplet_radius_1D = np.load('radius_1D.npy')

single_droplet_timestamps_2D = np.load('time_2D.npy')

single_droplet_radius_2D = np.load('radius_2D.npy')

single_droplet_timestamps_3D = np.load('time_3D.npy')

single_droplet_radius_3D = np.load('radius_3D.npy')
```

```python
AGM_t_max = single_droplet_timestamps_3D[-1][-1]

print(round(AGM_t_max, 2))

print()
```

### Plot Coarsening

```python
plt.style.use('default')

set_fontsize = 20

plt.rcParams['figure.constrained_layout.use'] = True

plt.figure(figsize=(5, 5))

for droplet in range(N_DROPLETS):

    plt.semilogx(single_droplet_timestamps_3D[droplet], single_droplet_radius_3D[droplet], 'darkorange', 
                 alpha = np.random.uniform(0.1, 0.2), linewidth = 3)
        
plt.axhline(y = 68.609, linestyle = '--', color = 'darkorange', alpha = 0.5, linewidth = 3)
    
##############################################################################################################
    
plt.xlim(1, AGM_t_max)

plt.ylim(1, 80)

plt.xlabel(r'$t \times \tau$', fontsize = set_fontsize)

plt.ylabel(r'$R/w$', fontsize = set_fontsize)

plt.tick_params(labelsize=set_fontsize)

plt.savefig('ActiveCoarsening3D.pdf')
```
