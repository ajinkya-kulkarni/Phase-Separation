### Analytical solution

#### Initialize the variables required

```python
r, theta = sym.symbols('r, theta', positive = True)

KF, KB, R, D, cEqIn, cEqOut, cInf, zeta = sym.symbols('KF, KB, R, D, c^{eq}_{in}, c^{eq}_{out}, c_{\infty}, \zeta',
                                                    positive = True)
```

```python
cInf = KF/(KF + KB)

cInf
```

```python
zeta = sqrt(D/(KF + KB))

zeta
```

######################################################################################################################

```python
C_1, C_2 = sym.symbols('C_1, C_2', real = True)
```

```python
c_in = cInf + C_1 * besselj(0, 1j * r / zeta) + C_2 * bessely(0, 1j * -r / zeta)

simplify(c_in)
```

### Verify the solution by plugging it in the equation

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_c_in_by_del_r = diff(c_in, r)
```

```python
del_c_in_by_del_theta = diff(c_in, theta)
```

```python
del_2_c_in_by_del_theta_2 = diff(del_c_in_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_c_in_by_del_r), r) + (1/(r*r))*del_2_c_in_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + KF*(1 - c_in) - KB*c_in) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

### Solve for $C_1$ and $C_2$

```python
c_in
```

######################################################################################################################

Use BC $\#2 \longrightarrow D\nabla c_{in} \vert_{r=R} = 0$

```python
Image("gradient2D.jpeg", width = 300)
```

```python
gradient = -D*(diff(c_in, r) + (1/r)*diff(c_in, theta))

gradient_at_centre = gradient.subs({r:0})

gradient_at_centre
```

```python
### hence C_2 is 0
```

Use BC $\#1 \longrightarrow c_{in} \vert_{r=R} = c^{eq}_{in}$

```python
eq1 = Eq(c_in.subs({r:R,C_2:0}) - cEqIn, 0)

eq1
```

```python
solved_C_1 = solve(eq1, C_1)[0]

solved_C_2 = 0
```

```python
solved_C_1, solved_C_2
```

```python
solved_c_in = c_in.subs({C_1: solved_C_1, C_2: solved_C_2})

solved_c_in
```

######################################################################################################################

### Verify the solution by plugging it in the equation (again)

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_solved_c_in_by_del_r = diff(solved_c_in, r)
```

```python
del_solved_c_in_by_del_theta = diff(solved_c_in, theta)
```

```python
del_2_solved_c_in_by_del_theta_2 = diff(del_solved_c_in_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_solved_c_in_by_del_r), r) + \
                    (1/(r*r))*del_2_solved_c_in_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + KF*(1 - solved_c_in) - KB*solved_c_in) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

######################################################################################################################

### Fluxes evaluated inside the droplet

```python
real_solved_c_in = re(solved_c_in.subs({KF:1e-5, KB:1e-4, D:1, cEqIn:(1+(2*0.0833)/R)}))

real_solved_c_in
```

```python
str(real_solved_c_in)
```

```python
Image("gradient2D.jpeg", width = 300)
```

```python
J_IN = -D*( diff(real_solved_c_in, r) + (1/r)*diff(real_solved_c_in, theta))

J_IN_at_surface = J_IN.subs({r:R, D:1})

simplify(J_IN_at_surface)
```

######################################################################################################################

######################################################################################################################

######################################################################################################################

```python
C_3, C_4 = sym.symbols('C_3, C_4', real = True)
```

```python
c_out = cInf + C_3 * besselj(0, 1j * r / zeta) + C_4 * bessely(0, 1j * -r / zeta)

simplify(c_out)
```

### Verify the solution by plugging it in the equation

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_c_out_by_del_r = diff(c_out, r)
```

```python
del_c_out_by_del_theta = diff(c_out, theta)
```

```python
del_2_c_out_by_del_theta_2 = diff(del_c_out_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_c_out_by_del_r), r) + (1/(r*r))*del_2_c_out_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + KF*(1 - c_out) - KB*c_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

### Solve for $C_3$ and $C_4$

```python
c_out
```

######################################################################################################################

Use BC $\#1 \longrightarrow c_{out} \vert_{r=L} = c_{\infty}$ Later set $L = \infty$

```python
L = sym.symbols('L', real = True)
```

```python
eq3 = Eq(c_out.subs({r:L}) - cInf, 0)

eq3
```

######################################################################################################################

Use BC $\#2 \longrightarrow c_{out} \vert_{r=R} = c^{eq}_{out}$

```python
eq4 = Eq(c_out.subs({r:R}) - cEqOut, 0)

eq4
```

```python
solved_C_3, solved_C_4 = list(solve((eq3, eq4), (C_3, C_4)).values())
```

```python
solved_c_out = c_out.subs({C_3:solved_C_3, C_4:solved_C_4})
```

######################################################################################################################

### Verify the solution by plugging it in the equation (again)

```python
simplify(solved_c_out.subs({r:L}))
```

```python
Image("laplacian2D.jpeg", width = 300)
```

```python
del_solved_c_out_by_del_r = diff(solved_c_out, r)
```

```python
del_solved_c_out_by_del_theta = diff(solved_c_out, theta)
```

```python
del_2_solved_c_out_by_del_theta_2 = diff(del_solved_c_out_by_del_theta, theta)
```

```python
polar_laplacian = (1/r)*diff((r*del_solved_c_out_by_del_r), r) + \
                    (1/(r*r))*del_2_solved_c_out_by_del_theta_2
```

```python
if (simplify(D * polar_laplacian + KF*(1 - solved_c_out) - KB*solved_c_out) != 0):
    
    raise ValueError("ERROR: Check the analytics again")
    
print()
```

######################################################################################################################

### Fluxes evaluated outside the droplet

```python
box_size = 1000

real_solved_c_out = re(solved_c_out.subs({KF:1e-5, KB:1e-4, D:1, cEqOut:(2*0.0833)/R, L:box_size*R}))

str(real_solved_c_out)
```

```python
Image("gradient2D.jpeg", width = 300)
```

```python
J_OUT = -D*( diff(real_solved_c_out, r) + (1/r)*diff(real_solved_c_out, theta))

J_OUT_at_surface = J_OUT.subs({r:R, D:1})
```

######################################################################################################################

######################################################################################################################

######################################################################################################################

```python
interface_velocity = (J_IN_at_surface - J_OUT_at_surface)/(1) # as cEqIn - cEqOut = 1

str(simplify(interface_velocity))
```

######################################################################################################################

######################################################################################################################

######################################################################################################################

### Solve it numerically

```python
max_radius = 35
```

```python
input_radius = np.linspace(0.5*max_radius, 1.5*max_radius, 50)

input_radius
```

```python
radius_array = []

for input_number in input_radius:
    
    radius_array.append([input_number, re(N(simplify(interface_velocity.subs({R:input_number})), 3)),
                         im(N(simplify(interface_velocity.subs({R:input_number})), 3))])

    print('Radius=' + str(round(input_number, 3)) + '; done')

print()

radius_array = np.asarray(radius_array)
```

```python
radius_array[:, 1]
```

```python
plt.plot(radius_array[:, 0], radius_array[:, 1], linewidth = 4)

plt.xlabel(r'R')

plt.ylabel('Droplet speed')

plt.axhline(y = 0, linewidth = 4, alpha = 0.2)

plt.xlim(min(input_radius), max(input_radius))
```

```python
index = np.argmin(np.abs(radius_array[:, 1] - 0))

stable_radius = round(radius_array[:, 0][index - 1], 4)

print('Numerical stable radius is ' + str(stable_radius))

print()
```

```python
from sympy.plotting import plot as symplot

symplot((real_solved_c_in.subs({R:stable_radius}), (r, 0, stable_radius)), 
        (real_solved_c_out.subs({R:stable_radius}), (r, stable_radius, 10*stable_radius)))
```
